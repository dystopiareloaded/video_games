```python
import sqlite3
import pandas as pd

# Load CSV file
df = pd.read_csv("video_game_sales.csv")

# Connect (or create) SQLite database
conn = sqlite3.connect("video_game_sales.db")
cursor = conn.cursor()

# Create the table if it doesn't exist
cursor.execute("""
CREATE TABLE IF NOT EXISTS video_game_sales (
    name TEXT,
    platform TEXT,
    year INTEGER,
    genre TEXT,
    publisher TEXT,
    global_sales REAL
)
""")

# Insert data into the table, replacing if it already exists
df.to_sql("video_game_sales", conn, if_exists="replace", index=False)

print("✅ CSV successfully imported into video_game_sales.db")

```

```python
# Query sample data
df = pd.read_sql_query("SELECT * FROM video_game_sales LIMIT 10;", conn)
```
